3. Dataset Overview

```python
def dataset_overview(df, name):
    print(f"\n{'='*50}")
    print(f"{name} OVERVIEW")
    print(f"{'='*50}")
    
    print("\nShape:", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())
    
    print("\nHead:")
    display(df.head())


dataset_overview(train, "TRAIN")
dataset_overview(test, "TEST")
dataset_overview(econ, "ECONOMIC DATA")
```

4. Schema and Data Types Validation

```python
def check_dtypes(df, name):
    print(f"\n{name} DATA TYPES")
    print(df.dtypes)

check_dtypes(train, "TRAIN")
check_dtypes(test, "TEST")
```

5. Missing Values Analysis

```python
def missing_values(df, name):
    print(f"\n{name} MISSING VALUES")
    
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values(ascending=False)
    
    if missing.empty:
        print("No missing values ✅")
    else:
        display(pd.DataFrame({
            "missing_count": missing,
            "missing_pct": (missing / len(df)) * 100
        }))

missing_values(train, "TRAIN")
missing_values(test, "TEST")
missing_values(econ, "ECONOMIC")
```

6. Duplicate & Key Integrity Checks

```python
def duplicate_checks(df, name):
    print(f"\n{name} DUPLICATE CHECKS")
    
    print("Full duplicates:", df.duplicated().sum())
    
    if "ID" in df.columns:
        print("Duplicate IDs:", df["ID"].duplicated().sum())

duplicate_checks(train, "TRAIN")
duplicate_checks(test, "TEST")
```

14. Initial Risk Flags

```python
print("\n🚨 INITIAL RISK FLAGS")

# Potential leakage
suspect_cols = [
    "Lender_portion_to_be_repaid"
]

for col in suspect_cols:
    if col in train.columns:
        print(f"⚠️ Potential leakage column: {col}")

# Check if repayment > amount (interest)
print("\nChecking repayment vs amount:")
check = (train["Total_Amount_to_Repay"] < train["Total_Amount"]).sum()
print("Repay < amount cases:", check)
```

```python
# ======================
# Data Type Corrections
# ======================

def fix_dtypes(df):
    df = df.copy()
    
    # Convert dates
    df["disbursement_date"] = pd.to_datetime(df["disbursement_date"], errors="coerce")
    df["due_date"] = pd.to_datetime(df["due_date"], errors="coerce")
    
    # Convert categoricals
    categorical_cols = ["country_id", "loan_type", "New_versus_Repeat"]
    
    for col in categorical_cols:
        df[col] = df[col].astype("category")
    
    return df


train = fix_dtypes(train)
test = fix_dtypes(test)

print("✅ Data types corrected")
```

**LOAN TYPE SUPPORT SIZE** 

Helps detect Rare categories causing fake high default rates

```python
loan_stats = train.groupby("loan_type")["target"].agg(["mean", "count"])
display(loan_stats.sort_values("mean", ascending=False))
```

🚨 CRITICAL INSIGHT

❗ High default rates = LOW SAMPLE SIZE

+ Type_20 → only 18 samples

+ Type_12 → only 4 samples

👉 These are statistical illusions, not real signals


If used directly:

+ Model will overfit rare categories
+ Poor generalization to Ghana

FINAL DECISION -> Group rare loan types

```python
rare_threshold = 100

loan_counts = train["loan_type"].value_counts()
rare_types = loan_counts[loan_counts < rare_threshold].index

# Add "Other" to categories first
train["loan_type"] = train["loan_type"].cat.add_categories(["Other"])
test["loan_type"] = test["loan_type"].cat.add_categories(["Other"])

# Now safely replace
train["loan_type_grouped"] = train["loan_type"].replace(rare_types, "Other")
test["loan_type_grouped"] = test["loan_type"].replace(rare_types, "Other")
```

# 📊 Data Validation & Sanity Check — Summary Report

## 1. 🎯 Objective

The purpose of this notebook was to conduct a comprehensive data validation and sanity check of all datasets provided for the loan default prediction task. This step ensures:

* Data integrity and consistency
* Identification of potential data leakage
* Understanding of dataset structure and relationships
* Early detection of modeling risks
* Establishment of a reliable foundation for feature engineering and modeling

---

## 2. 📁 Datasets Overview

The project utilizes three primary datasets:

* **Train Dataset**: Contains labeled loan records with default outcomes
* **Test Dataset**: Contains unlabeled loan records for prediction
* **Economic Indicators Dataset**: Macroeconomic variables across countries and years

---

## 3. ✅ Key Validation Findings

### 3.1 Missing Values

* **Train Dataset**: No missing values
* **Test Dataset**: No missing values
* **Economic Dataset**:

  * Significant missingness (up to ~48%)
  * Higher missing rates in recent years and certain countries (e.g., Ghana)

📌 **Implication**:

* Core datasets are clean and ready for modeling
* Economic data requires careful handling (imputation or missing indicators)

---

### 3.2 Duplicate Checks

* No duplicate rows detected
* No duplicate IDs detected

📌 **Implication**:

* Dataset integrity is strong
* No need for deduplication

---

### 3.3 Data Types

* Numerical and categorical columns are correctly identified
* Date columns were initially stored as strings and converted to datetime
* Categorical columns require explicit handling (`category` dtype)

📌 **Implication**:

* Data types are now aligned for efficient processing and modeling

---

## 4. 🧠 Structural Insights

### 4.1 Entity Relationships

* **Customers**: ~6,500
* **Loans**: ~66,000
* **Lenders**: 4

#### Key Observations:

* Customers have multiple loans (mean ≈ 10 loans per customer)
* Loans are typically funded by a single lender (multi-lender cases are rare)

📌 **Implication**:

* Strong **customer-level behavioral signal**
* Limited importance of lender aggregation

---

## 5. ⚠️ Target Variable Analysis

* Default rate: **~1.83%**
* Highly imbalanced dataset

📌 **Implication**:

* Accuracy is not a reliable metric
* Requires:

  * Stratified validation
  * Class weighting
  * Threshold optimization

---

## 6. 🚨 Data Leakage Detection

### Identified Risk:

* `Lender_portion_to_be_repaid`

### Evidence:

* Strong statistical separation between default and non-default classes

📌 **Conclusion**:

* This feature likely encodes post-outcome information
* **Will be removed from modeling**

---

## 7. 📊 Feature-Level Insights

### 7.1 Loan Type

* Some loan types show extremely high default rates
* These categories have very low sample sizes

📌 **Implication**:

* High variance / unreliable signals
* Rare categories must be grouped

---

### 7.2 New vs Repeat Loans

* New loans: ~19.9% default rate
* Repeat loans: ~1.69% default rate

📌 **Implication**:

* One of the strongest predictive features
* Reflects borrower trust and history

---

### 7.3 Customer Behavior

* Most customers never default
* Some customers consistently default

📌 **Implication**:

* Strong opportunity for **customer-level feature engineering**
* Must ensure time-aware computation to avoid leakage

---

### 7.4 Loan Amount

* Larger loans show higher default rates
* Distribution is highly skewed with extreme outliers

📌 **Implication**:

* Consider log transformations or tree-based models

---

### 7.5 Loan Duration

* Longer durations are associated with higher default rates

📌 **Implication**:

* Key risk indicator
* Strong predictive feature

---

### 7.6 Temporal Patterns

* Data heavily concentrated in 2022
* Smaller sample sizes in other years
* Some years (e.g., 2021) show inflated default rates due to low sample size

📌 **Implication**:

* Time-based bias exists
* Temporal features must be handled carefully

---

## 8. 🌍 Distribution Shift (Train vs Test)

### Observed Differences:

* **Duration**:

  * Train mean ≈ 8.5 days
  * Test mean ≈ 13.7 days

📌 **Implication**:

* Significant distribution shift
* Model must generalize beyond training distribution

---

## 9. 🌍 Economic Data Assessment

* Contains 9 macroeconomic indicators
* Varying levels of missingness by country:

  * Kenya: low missingness (~5%)
  * Ghana: high missingness (~37%)

📌 **Implication**:

* Useful for domain generalization
* Requires:

  * Careful merging
  * Missing value handling strategy

---

## 10. ⚠️ Key Risks Identified

1. **Severe Class Imbalance**
2. **Potential Data Leakage**
3. **Distribution Shift (Train vs Test)**
4. **Sparse and Noisy Categories**
5. **Temporal Bias**
6. **Economic Data Missingness**

---

## 11. ✅ Strategic Decisions

Based on validation findings, the following decisions are made:

* ❌ Drop leakage-prone features:

  * `Lender_portion_to_be_repaid`

* 🔧 Group rare categorical values:

  * `loan_type`

* 🧠 Prioritize customer-level features:

  * Behavioral and historical patterns

* ⚖️ Handle class imbalance:

  * Use F1-optimized strategies

* 🔄 Use grouped validation:

  * `GroupKFold` with `customer_id`

* 🌍 Incorporate economic indicators cautiously

---
